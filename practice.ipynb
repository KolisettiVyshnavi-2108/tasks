```python
import numpy as np
from scipy.io import loadmat

# Load data
main = loadmat("values2.mat")["values2"]
sub = np.loadtxt("submatrix_5.csv", delimiter=",")

print("Main matrix shape:", main.shape)
print("Total values in main matrix:", main.size)
print("Submatrix shape:", sub.shape)
print("Total values in submatrix:", sub.size)

# Find submatrix index
sub_matrix_index = -1
for i in range(main.shape[0]):
    if np.array_equal(main[i], sub):
        sub_matrix_index = i
        break

if sub_matrix_index == -1:
    print("Submatrix NOT found")
    exit()

print("Submatrix_index:", sub_matrix_index)

# Increase values by 50%
after = sub * 1.5

# ðŸ”¹ Manhattan Distance function
def manhattan_distances(target, main, skip_index=None):
    distances = []
    for i in range(main.shape[0]):
        if i == skip_index:
            distances.append(np.inf)
        else:
            distances.append(np.sum(np.abs(main[i] - target)))
    return np.array(distances)

# Calculate distances
distances_after = manhattan_distances(after, main, skip_index=sub_matrix_index)

# Find closest match
after_index = int(np.argmin(distances_after))

print("AFTER 50% HIGHER DEVIANT closest index:", after_index)
print("Manhattan distance to closest submatrix:", distances_after[after_index])
```

```python
import numpy as np
from scipy.io import loadmat

main = loadmat("values2.mat")["values2"]
sub = np.loadtxt("submatrix_5.csv", delimiter=",")

print("Main matrix shape:", main.shape)
print("Total values in main matrix:", main.size)
print("Submatrix shape:", sub.shape)
print("Total values in submatrix:", sub.size)

sub_matrix_index = -1
for i in range(main.shape[0]):
    if np.array_equal(main[i], sub):
        sub_matrix_index = i
        break

if sub_matrix_index == -1:
    print("Submatrix NOT found")
    exit()

print("Submatrix_index:", sub_matrix_index)

after = sub * 1.5

# ðŸ”¹ Mean Absolute Difference function
def mean_absolute_difference(target, main, skip_index=None):
    distances = []
    n = target.size   # total elements
    for i in range(main.shape[0]):
        if i == skip_index:
            distances.append(np.inf)
        else:
            mad = np.sum(np.abs(main[i] - target)) / n
            distances.append(mad)
    return np.array(distances)

# Calculate distances
distances_after = mean_absolute_difference(after, main, skip_index=sub_matrix_index)

after_index = int(np.argmin(distances_after))

print("AFTER 50% HIGHER DEVIANT closest index:", after_index)
print("Mean Absolute Difference to closest submatrix:", distances_after[after_index])
```
